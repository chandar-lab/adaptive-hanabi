```python
def compute_metric(cross_play_matrix):
    cross_play_matrix = cross_play_matrix / 25
    diag_idx = np.eye(cross_play_matrix.shape[0],dtype=bool)
    diversity = -np.mean(cross_play_matrix[~diag_idx]) + 1
    
    strength = np.mean(cross_play_matrix.max(axis=0))
    return diversity.round(2), strength.round(2)
```

```python
learner_labels = {'vdn-op1-lstmlayers1-seed100': 'VDN + OP', 'vdn-op0-lstmlayers1-seed100': 'VDN',
           'icml_OBL5/OFF_BELIEF1_SHUFFLE_COLOR0_LOAD1_BZA0_BELIEF_a': 'OBL', 'iql-op0-lstmlayers1-seed100': 'IQL',
           'iql-op1-lstmlayers1-seed100': 'IQL + OP'}
```

```python
import copy
learners = ['iql-op0-lstmlayers1-seed100', 'iql-op1-lstmlayers1-seed100', 'vdn-op0-lstmlayers1-seed100',
            'vdn-op1-lstmlayers1-seed100', 'icml_OBL5/OFF_BELIEF1_SHUFFLE_COLOR0_LOAD1_BZA0_BELIEF_a']
print(f"medium: {partners_1}")
print(f"low: {partners_3}")
print(f"high: {partners_2}")
print(f"learners: {learners}")
print("**********")
all_partners = list(set(learners + partners_1 + partners_2 + partners_3))

partners1_learners = list(set(partners_1 + learners))
print(f"all partners: {set(all_partners)}")
print("**********")

partners_2_copy = copy.deepcopy(partners_2)
for p in partners1_learners:
    if p in partners_2_copy:
        partners_2_copy.remove(p)

print(f"Low diversity Agents: {set(partners_2)}")

print(f"New Low diversity Agents: {set(partners_2_copy)}")
print("**********")

partners12_learners = list(set(partners_1 + learners + partners_2))

for p in partners12_learners:
    if p in partners_3:
        partners_3.remove(p)

print(f"Low diversity Agents: {set(partners_2)}")
print(f"New High diversity Agents: {set(partners_3)}")
```
